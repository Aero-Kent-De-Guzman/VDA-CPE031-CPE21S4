```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as num

heart_2022_A = pd.read_csv('heart_2022_no_nans.csv')
heart_2022_B = pd.read_csv('heart_2022_with_nans.csv')
Df1 = pd.DataFrame(heart_2022_A)
Df2 = pd.DataFrame(heart_2022_B)
Df1
```

# **QUESTION 1**  : On what degree is the data gathered useful when it comes to utilization as reference when it comes to figuring out the variables for a heart disease?

```python
HeartDisease_noNaN = Df1['GeneralHealth'].value_counts()
HeartDisease_withNaN = Df2['GeneralHealth'].value_counts()
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].bar(HeartDisease_noNaN.index, HeartDisease_noNaN.values)
axes[0].set_title('No NaN')
axes[1].bar(HeartDisease_withNaN.index, HeartDisease_withNaN.values)
axes[1].set_title('With NaN')

fig.suptitle("Data Subjects Health Condition")
plt.show()
```

## **ANSWER** :

Based on the data visualized in the graph, both data with either containing values of NULL or NaN and values containing randomized responses in order to fill in those Empty values both contains data that leans more on most of the respondents being a good condition. From that response we can say that the data is not viable to be utilized when it comes to understanding either the frequency or the prominent cause of a heart disease simply due to the fact that not even 50% of the respondents are viable to be used as a reference since their heath condition are good enough that there is either no need further undertake various test or simply conclude the disease as hereditary since they are already healthy.

Still, despite not even half of the respondents being viable and thus does not meet the critea needed to be used as a reference/data, the synthesized output can still be utilized in other ways such as understanding the relationship between the other variables instead like figuring out the relationship between the Body Mass Index (BMI) towards a person sleep hours with physical activities being utilzed additionally as well.

# **QUESTION 2** : Based on the data gathered, how much is it representable as a whole towards the society?

```python
Sex_noNaN = Df1['Sex'].value_counts()
Sex_withNaN = Df2['Sex'].value_counts()
fig, axes = plt.subplots(1, 2)

axes[0].pie(Sex_noNaN, labels=Sex_noNaN.index, autopct='%1.1f%%')
axes[0].set_title('No NaN')
axes[1].pie(Sex_withNaN, labels=Sex_withNaN.index, autopct='%1.1f%%')
axes[1].set_title('With NaN')

fig.suptitle("Data Subjects Gender (Sex)")
plt.show()
```

## **ANSWER** :

Based on the visualized data, both data with either containing values of NULL or NaN and values containing randomized responses in order to fill in those Empty values both contains data that are near evenly spread throughout the respondents. Due to this fact, we can conclude with certainty that the data is indeed representable as a whole for the society since both individuals with varying physical shapes/capabilities and attributes are both utilized in gathering the data. Still, the data merely gathered their responses on a specific area in the world which is some state in the United States which could obscure the certainty and condition of the data gathered which can make it a gap in the data due to the fact their living conditions(surrounding environment)can very much affect their health as well.
