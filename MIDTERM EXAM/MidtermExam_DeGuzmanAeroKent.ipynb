## **Question 1**

```python
HeartDisease = df['HeartDisease'].value_counts()
plt.pie(HeartDisease, labels=HeartDisease.index, autopct='%1.1f%%')
plt.title("Data Subjects Heart Condition")
plt.show()
```

### **Answer**
While the data does indeed aid in understanding to figure out whether or not a variable affects their hearts condition for the better or worse, it is not great due to the fact that in order to truly understand what are the leading causes or variables that determine or primarily affects whether an individual can contact a heart disease should be equal or more than 50%. In the visualized data, only 8.6% of the respondents have really provided valuable data due to the fact they already met the criteria which is to have a heart disease, gathering the additional data needed which is their health condition is the only one next. But because only few percentage of the overall respondents are viable to be used in the study, the data output that was synthesized in the data cannot be used when it comes to figuring out the health conditions that affects an individual.

To answer it shortly, the data gathered is only useful in a small degree since the data is not representable for the individuals that have a heart disease where it should be at least 50%. Still, the data can still be used in other topics such as what areas in an individuals health should be prioritarized in order to maintain a healthy life/heart.
