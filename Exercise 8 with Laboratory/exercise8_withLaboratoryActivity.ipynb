# **Matplotlib Pyplot**

## **Matplot Labels**

```python
xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])

plt.plot(xpoints,ypoints)

plt.title("Distance Travelled Everyday")
plt.xlabel("Distance in km")
plt.ylabel("Speed in km/s")
plt.show()
```

### with color

```python
xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])

plt.plot(xpoints,ypoints)

plt.title("Distance Travelled Everyday", color = 'yellow')
plt.xlabel("Distance in km")
plt.ylabel("Speed in km/s")
plt.show()
```

### changing label alignment

```python
xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])

plt.plot(xpoints,ypoints)

plt.title("Distance Travelled Everyday", loc = 'left')
plt.xlabel("Distance in km")
plt.ylabel("Speed in km/s")
plt.show()
```

### Creating grid

```python
xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])

plt.plot(xpoints,ypoints)

plt.title("Distance Travelled Everyday")
plt.xlabel("Distance in km")
plt.ylabel("Speed in km/s")
plt.grid()
plt.show()
```

### 1-axis grid

```python
xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])

plt.plot(xpoints,ypoints)

plt.title("Distance Travelled Everyday")
plt.xlabel("Distance in km")
plt.ylabel("Speed in km/s")
plt.grid(axis = 'y')
plt.show()
```

### grid styles

```python
xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])

plt.plot(xpoints,ypoints)

plt.title("Distance Travelled Everyday")
plt.xlabel("Distance in km")
plt.ylabel("Speed in km/s")
plt.grid(axis = 'x', linestyle = '--', linewidth = 0.50)
plt.show()
```

### subplots

```python
#plot 1
xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(xpoints,ypoints)

#plot 2
xpoints = np.array([1,2,6,8])
ypoints = np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.plot(xpoints,ypoints)

plt.grid(axis = 'x', linestyle = '--', linewidth = 0.50)
plt.show()
```

## **Pie Chart**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
g = np.array([35,25,25,15])
grader = ["Grade 1=35%","Grade 2=25%","Grade 3=25%","Grade 4=15%"]
plt.pie(g, labels = grader)
plt.show()
```

### Creating the Legend and its location

```python
g = np.array([35,25,25,15])
grader = ["Grade 1=35%","Grade 2=25%","Grade 3=25%","Grade 4=15%"]
plt.pie(g, labels = grader)
plt.legend(labels = grader, loc = "upper left")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,25,30]
plt.plot(x,y)
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,25,30]
plt.scatter(x,y)
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt

data = [1,2,2,3,3,3,4]
plt.hist(data)
plt.title("Histrogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C']
values = [5,7,3]
plt.bar(categories, values)
plt.title("Bar Chart Example")
plt.xlabel("Caterogies")
plt.ylabel("Values")
plt.show()
```

```python
import matplotlib.pyplot as plt

data = [[1.5]*10 + [2]*10 + [3]*10]
plt.boxplot(data)
plt.title("Box Plot Example")
plt.ylabel("Values")
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C']
sizes = [40,30,30]

plt.pie(sizes, labels=labels)
plt.title("Pie Chart Example")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Creating the  Dataframe

data = pd.read_csv('user_behavior_dataset.csv')
df = pd.DataFrame(data)

df["User Behavior Class"].values
```

```python
data1 = df["User Behavior Class"].values

plt.hist(data1)
plt.title("User Behavior Class")
plt.xlabel("Behavior Class")
plt.ylabel("Sample Size")
plt.show()
```

```python
data2 = df["App Usage Time (min/day)"].values

plt.boxplot(data2, showmeans = 'true')
plt.title("App Usage Time (min/day)")
plt.ylabel("Frequency")
plt.show()
```

```python
data3 = df["Age"].values

plt.hist(data3)
plt.title("Age")
plt.xlabel("Age Group")
plt.ylabel("Sample Size")
plt.show()
```

```python
data4 = df["Screen On Time (hours/day)"].values

plt.boxplot(data4, showmeans = 'true')
plt.title("Screen On Time (hours/day)")
plt.ylabel("Frequency")
plt.show()
```
